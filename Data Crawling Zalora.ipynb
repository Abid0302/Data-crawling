```python
browser.get('https://www.zalora.co.id/c/sports/sports-pria/c-10283?from=header&csa=bbcmsports')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
# mengambil satu data
soup.find('div', class_='overflow-hidden text-ellipsis whitespace-nowrap text-sm').text.strip()
```

# Looping ambil semua data

```python
nama_product = soup.find_all('div', class_='overflow-hidden text-ellipsis whitespace-nowrap text-sm')
daftar_nama = []
for tiap_nama_product in nama_product:
    daftar_nama.append(tiap_nama_product.text.strip())
```

```python
daftar_nama
```

```python
harga = soup.find_all('div', class_='mt-1 flex flex-col flex-wrap items-start gap-1 whitespace-nowrap text-base tablet:flex-row tablet:items-center')
daftar_harga = []
for tiap_harga in harga:
    daftar_harga.append(tiap_harga.text.strip())
```

```python
daftar_harga
```

```python
data = {
    'nama_produk' : daftar_nama,
    'harga' : daftar_harga
}
```

```python
file = pd.DataFrame(data)
```
